```python
etf = pd.read_csv(r'C:\Workbook\Funds\ETF2.csv', encoding = 'utf-8')
```

```python
# Establishing Local DB connection

db_connection = sqlite3.connect(r'C:\Workbook\Funds\etf.db.sqlite')
c = db_connection.cursor()
# Save these two dataframes as a table in sqlit
# Save these two dataframes as a table in sqlite
etf.to_sql('etf', con=db_connection,if_exists = 'replace', index = False)
```

```python
# Count number of funds per FM
result1='SELECT  COUNT(fund_extended_name) AS Num_fund,fund_family FROM etf GROUP BY fund_family ORDER BY Num_fund DESC'
sql.read_sql(result1, con=db_connection)
```

```python
# AVG PE Ratio per Fund Family
result2 ='SELECT AVG (price_earnings_ratio) AS PE_AVG,fund_family, investment_type FROM etf GROUP BY fund_family ORDER BY PE_AVG DESC'
sql.read_sql(result2, con=db_connection)
```

```python
# AVG Funds_Returns Ratio per Fund Family
result3 ='SELECT AVG (fund_return_10years) AS AVG_PERF,fund_family FROM etf GROUP BY fund_family  ORDER BY fund_return_10years DESC'
sql.read_sql(result3, con=db_connection)
```

```python
# AVG Funds_Returns Ratio per Fund Family
result4 ='SELECT investment_type,AVG (fund_return_10years) as AVG_FUND_RETURN,price_earnings_ratio, fund_family FROM etf WHERE price_earnings_ratio <= 15 GROUP BY fund_family  ORDER BY  price_earnings_ratio DESC'
sql.read_sql(result4, con=db_connection)
```

```python
# Best Funds
result5 ='SELECT investment_type, MAX (fund_return_10years) as BEST_FUND_RETURN,price_earnings_ratio, fund_family,fund_extended_name FROM etf GROUP BY fund_family ORDER BY BEST_FUND_RETURN DESC'
sql.read_sql(result5, con=db_connection)
```

```python
# Worst Funds
result6 ='SELECT investment_type, MIN (fund_return_10years) as WORST_FUND_RETURN,price_earnings_ratio, COUNT(fund_extended_name),fund_family,fund_extended_name FROM etf GROUP BY fund_family ORDER BY WORST_FUND_RETURN ASC'
sql.read_sql(result6, con=db_connection)
```

```python
# Worst Fund
result7 ='SELECT investment_type, MIN (fund_return_10years) as WORST_FUND_RETURN,price_earnings_ratio,fund_family,fund_extended_name FROM etf WHERE fund_extended_name ="VanEck Vectors Egypt Index ETF"'
sql.read_sql(result7, con=db_connection)
```

```python
# 5 BEST Funds
result8 ='SELECT fund_extended_name, fund_return_10years as BEST_FUND_RETURN, price_earnings_ratio, fund_family FROM etf ORDER BY BEST_FUND_RETURN DESC LIMIT 5'
sql.read_sql(result8, con=db_connection)
```
